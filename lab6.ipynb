```julia
using Plots
using DifferentialEquations
```

```julia
a = 0.01;
b = 0.02;
N =  20000;
I0 = 99;
R0 = 5;
S0 = N - I0 - R0;
```

```julia
# случай 1
    function inf1(dx,x,p,t)
        dx[1] = 0
        dx[2] = -b*x[2]
        dx[3] = b*x[2]
    end
 x0 = [S0, I0, R0]
 tspan = (0, 90)
 prob1 = ODEProblem(inf1, x0, tspan)
 sol1 = solve(prob1, timeseries_steps = 0.01);
```

```julia
p1 = plot(sol1, 
    label = ["S(t)" "I(t)" "R(t)"],
    title = "Модель заражания №1",
    titlefontsize = 15)
```

```julia

    function inf2(dx,x,p,t)
        dx[1] = -a*x[1]
        dx[2] = a*x[1] - b*x[2]
        dx[3] = b*x[2]
    end
 x0 = [S0, I0, R0]
 tspan = (0, 1000)
 prob2 = ODEProblem(inf2, x0, tspan)
 sol2 = solve(prob2, timeseries_steps = 0.01);
```

```julia
p2 = plot(sol2, 
    label = ["S(t)" "I(t)" "R(t)"],
    title = "Модель заражания №2",
    titlefontsize = 10)
```
